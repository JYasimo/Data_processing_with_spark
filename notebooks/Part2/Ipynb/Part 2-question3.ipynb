3.For each taxi colour (yellow and green), each pair of pick up and drop off locations (use boroughs not the id), each month, each day of week and each hours:

What was the total number of trips?

What was the average distance?

What was the average amount paid per trip (using total_amount)?

What was the total amount paid (using total_amount)?


```
#reading combined_df from dbfs
from pyspark.sql.functions import lit
from pyspark.sql import SparkSession

df_path= "/dbfs/mnt/bde2/combined_df"
combined_df= spark.read.parquet(df_path)
```

```
#temp table combined_df
temp_table_name = 'combined_df'
combined_df.createOrReplaceTempView(temp_table_name)
```

```
%sql
CREATE OR REPLACE TEMP VIEW question3 AS--create temp table with needed columns
SELECT
  year,
  month,
  hour,
  weekday,
  Borough_pu,
  Borough_do,
  trip_distance_km,
  total_amount,
  colour
FROM combined_df;
```

```
%sql
SELECT  --Select and aggregate data related to taxi trips, categorizing by various attributes.
    CASE WHEN colour = 1 THEN 'yellow' ELSE 'green' END AS taxi_color,-- Assign 'yellow' if colour is 1, else 'green'
    Borough_pu AS pickup_location,-- Select the pickup location (Borough).
    Borough_do AS dropoff_location,-- Select the dropoff location (Borough).
    CASE month
        WHEN 1 THEN 'January'
        WHEN 2 THEN 'February'
        WHEN 3 THEN 'March'
        WHEN 4 THEN 'April'
        WHEN 5 THEN 'May'
        WHEN 6 THEN 'June'
        WHEN 7 THEN 'July'
        WHEN 8 THEN 'August'
        WHEN 9 THEN 'September'
        WHEN 10 THEN 'October'
        WHEN 11 THEN 'November'
        WHEN 12 THEN 'December'
    END AS month_name,-- Convert the month number to its name.
    CASE weekday
        WHEN 1 THEN 'Monday'
        WHEN 2 THEN 'Tuesday'
        WHEN 3 THEN 'Wednesday'
        WHEN 4 THEN 'Thursday'
        WHEN 5 THEN 'Friday'
        WHEN 6 THEN 'Saturday'
        WHEN 7 THEN 'Sunday'
    END AS day_of_week,-- Convert the weekday number to its name.
    hour AS hour,-- Select the hour of the day.
    COUNT(*) AS total_trips,-- Count the number of trips in each category.
    ROUND(AVG(trip_distance_km), 2) AS avg_distance,-- Calculate and round the average trip distance.
    ROUND(AVG(total_amount), 2) AS avg_amount_paid,-- Calculate and round the average total amount paid.
    ROUND(SUM(total_amount), 2) AS total_amount_paid-- Calculate and round the total amount paid.
FROM
    question3 -- The source table 'question3'.
GROUP BY
    taxi_color,
    pickup_location,
    dropoff_location,
    month_name,
    day_of_week,
    hour-- Group the data by these attributes for aggregation.
ORDER BY
   month_name DESC, day_of_week, hour ASC, taxi_color, pickup_location, dropoff_location;
```
